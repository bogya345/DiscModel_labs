# Determine the fertility rate for women in the age category “20- ... -39”

```python
# data about 
# path2 = './storage/WPP2019_FERT_F01_BIRTHS_BOTH_SEXES.xlsx'
path2 = './storage/age_data_2005.xls'
sheet2m = 'm; 1950-2005, estimates'
sheet2f = 'f; 1950-2005, estimates'
```

```python
cols2 = [
    'Index', 'Variant', 'Area', 'Notes',
    'Country_code', 'Reference_date',
    '0 - 4', '5 - 9', '10 - 14', '15 - 19',
    '20 - 24', '25 - 29', '30 - 34', '35 - 39',
    '40 - 44', '45 - 49', '50 - 54', '55 - 59', '60 - 64',
    '65 - 69', '70 - 74', '75 - 79', '80 - 84', '85 - 89', '90 - 94',
    '95 - 99', '100+'
]
cols2_ext_period = [ 'Sex',
    '0 - 4', '5 - 9', '10 - 14', '15 - 19',
    '20 - 24', '25 - 29', '30 - 34', '35 - 39',
    '40 - 44', '45 - 49', '50 - 54', '55 - 59', '60 - 64',
    '65 - 69', '70 - 74', '75 - 79', '80 - 84'
    # ,'85 - 89', '90 - 94', '95 - 99', '100+'
]
cols2_period = [
    '0 - 4', '5 - 9', '10 - 14', '15 - 19',
    '20 - 24', '25 - 29', '30 - 34', '35 - 39',
    '40 - 44', '45 - 49', '50 - 54', '55 - 59', '60 - 64',
    '65 - 69', '70 - 74', '75 - 79', '80 - 84'
    # ,'85 - 89', '90 - 94', '95 - 99', '100+'
]
```

```python
df2m = pd.read_excel(path2, sheet_name=sheet2m)
df2m.columns = cols2
df2m.dropna(subset=['Index', 'Variant', 'Area'], inplace=True)
df2m = df2m.iloc[1: , :]
df2m['Sex'] = 'm'
df2m
```

```python
df2f = pd.read_excel(path2, sheet_name=sheet2f)
df2f.columns = cols2
df2f.dropna(subset=['Index', 'Variant', 'Area'], inplace=True)
df2f = df2f.iloc[1: , :]
df2f['Sex'] = 'f'
df2f
```

```python
df2 = df2m.append(df2f)
df2
```

```python
d2 = df2[ df2['Country_code']==choosenCountryCode ][cols2_ext_period]
d2
```

```python
def fertility(amount_births, amount_women):
    return ( amount_births / amount_women ) * 1e3
```

```python
popularities = pd.DataFrame([d2[cols2_period].sum()], columns=cols2_period)
popularities
```

```python
vals = popularities.values[0]

tmp = []

for i in range(1, len(vals)):
    prev_births = vals[i-1] * 1e3
    amount_women = df2[ df2['Sex']=='m' ][popularities.columns[i-1]].sum()
    # print(amount_women, popularities.columns[i])

    births = vals[i]
    
    x = fertility(births, amount_women)
    tmp.append(x)
    continue

print(tmp)
```

```python
r2 = pd.DataFrame([tmp], columns=cols2_period[1:])
r2
```
