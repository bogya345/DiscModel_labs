```python
def calc_surv_rates(data, labels):
    surv_rates = pd.DataFrame([], columns=['Referenced_date', *labels])
    # surv_rates_m = pd.DataFrame([], columns=col2)
    # surv_rates

    tmp_survs = []
    for row_ind in range(1, data.shape[0]):

        year_cur_data = data.iloc[row_ind-1]
        year_next_data = data.iloc[row_ind]
        coefs = []
        for col_ind in range(1, len(cols_period)):
            cur_alive = year_cur_data[cols_period[col_ind-1]]
            next_alive = year_next_data[cols_period[col_ind]]
            # print(cur_alive, next_alive)

            coefs.append(cur_alive/next_alive)

            continue
        # print('===============', coefs)

        tmp_rec = pd.DataFrame([[int(data.iloc[row_ind].Reference_date),
                                # 0,
                                *coefs]], columns=['Referenced_date', *labels])
        # print(tmp_rec)
        
        surv_rates = surv_rates.append(tmp_rec)

        continue

    return surv_rates
```

# Predict the change in the country's population and demographic profile for 100 years and compare with existing prediction!

```python
# 0-4 перейдут в 5-9 - с коеф.
```

```python
model_cols = cols_ext_period
print(model_cols)
```

```python
# args - 
def model(coef_fertility, coef_bg_ratio, surv_rate_m, surv_rate_f, prev_record_m, prev_record_f, period_labels=cols_period):
    # res = pd.DataFrame([], columns=model_cols)
    # res.Reference_date = start_year
    
    resm = []
    resf = []
    
    resm.append(prev_record_m[0]+5)
    resf.append(prev_record_f[0]+5)

    resm.append('m')
    resf.append('f')

    tmpm = []
    for prev_period, rate in zip(prev_record_m[2:], surv_rate_m):
        alive_predict_m = (prev_period * rate)
        
        tmpm.append(alive_predict_m)
        continue

    tmpf = []
    for prev_period, rate in zip(prev_record_f[2:], surv_rate_f):
        alive_predict_f = (prev_period * rate)
        
        tmpf.append(alive_predict_f)
        continue

    women_amount = sum(tmpf[5:9])

    birthsm = (women_amount * coef_fertility) * coef_bg_ratios[0]
    resm.append(birthsm)

    birthsf = (women_amount * coef_fertility) * coef_bg_ratios[1]
    resf.append(birthsf)

    [resm.append(i) for i in tmpm]
    [resf.append(i) for i in tmpf]

    return resm, resf
```

```python
pred_male, pred_female = model(
    coef_fertility=coef_fertility,
    coef_bg_ratio=coef_bg_ratios,
    surv_rate_m=surv_rate_m.values[0], surv_rate_f=surv_rate_f.values[0],
    prev_record_m = dmal.iloc[-1].values, prev_record_f= dfem.iloc[-1].values
    # ,period_labels=labels
)

print(pred_male, pred_female)
```
