# Determine the fertility rate for women in the age category “20- ... -39”

```python
# data about 
path2 = './storage/WPP2019_FERT_F01_BIRTHS_BOTH_SEXES.xlsx'
```

```python
cols2 = [
    'Index',
    'Variant', 'Area', 'Notes', 'Country_code', 'Type', 'Parent_code',
    '1950-1955', '1955-1960', '1960-1965', '1965-1970',	'1970-1975',
    '1975-1980', '1980-1985', '1985-1990', '1990-1995',	'1995-2000',
    '2000-2005', '2005-2010', '2010-2015', '2015-2020'
]
periods = ['1950-1955', '1955-1960', '1960-1965', '1965-1970',	'1970-1975',
           '1975-1980', '1980-1985', '1985-1990', '1990-1995',	'1995-2000',
           '2000-2005', '2005-2010', '2010-2015', '2015-2020']
print(len(cols2))
```

```python
df2 = pd.read_excel(path2)
df2.columns = cols2
df2.dropna(subset=['Index', 'Variant', 'Area'], inplace=True)
df2 = df2.iloc[1: , :]
df2
```

```python
d2 = df2[ df2['Country_code']==choosenCountryCode ][periods]
d2
```

```python
def fertility(amount_births, amount_women):
    return ( amount_births / amount_women ) * 1e3
```

```python
# let assume that in each period relation male\female is 40\60
amount_women_coef = .6
```

```python
vals = d2.values[0]

tmp = []

for i in range(1, len(vals)):
    prev_births = vals[i-1] * 1e3
    amount_women = prev_births * amount_women_coef

    births = vals[i]
    
    x = fertility(births, amount_women)
    tmp.append(x)
    continue

print(tmp)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(0, figsize=(20,5))
b = plt.bar(d2.columns[1:], height=tmp, width=0.1)
plt.grid()
plt.bar_label(b)
plt.show()
```
